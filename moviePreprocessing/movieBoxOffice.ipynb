```python
# Initial imports
import pandas as pd
```

## Box-office boxoffice from the web

```python
# Scrap boxoffice for all years, and get an unique dataframe

# Scrap boxoffices for the given year
def scrap_boxoffice_total(year):
    df_box = pd.read_html(f"https://www.boxofficemojo.com/year/{year}/?grossesOption=totalGrosses")[0]
    df_box["Year"] = year
    return df_box

# Concatenate each year's boxoffice
boxoffice = pd.concat([scrap_boxoffice_total(year) for year in range(2015, 2021)])
print(boxoffice.shape)
```

```python
boxoffice
```

```python
# Format day
boxoffice["Release Date"] = boxoffice["Open"] + ", " + boxoffice["Year"].astype(int).astype(str)

# Map "Release Date" and "date" to a date format
boxoffice["Release Date"] = pd.to_datetime(boxoffice["Release Date"])

boxoffice
```

```python
# Export the full dataset
boxoffice.to_csv("../data/boxoffice.csv")
```
